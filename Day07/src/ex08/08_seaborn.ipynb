```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
query = """
SELECT 
    uid,
    timestamp,
    numTrials
FROM checker
WHERE 
    uid LIKE 'user_%'
    AND status = 'ready'
    AND labname = 'project1'
"""

df = pd.read_sql_query(query, conn, parse_dates=['timestamp'])
df
```

```python
plt.figure(figsize=(15,10))
sns.set_style("darkgrid", {'axes.faccecolor':'lightgrey'})
sns.lineplot(
    data=df,
    x='timestamp',
    y='numTrials',
    hue='uid',
    palette='tab20c',
    linewidth=3,
    sort=True,
)
plt.legend(bbox_to_anchor=(1, 0.9), loc=2)
plt.grid(True)
plt.suptitle('Dynamic of commits per user in project1', fontsize=30, y= 0.8)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)


plt.xlabel('timestamp',fontsize=15)
plt.ylabel('numTrials',fontsize=15)

plt.xlim(df['timestamp'].min(), df['timestamp'].max())

plt.show()
```
