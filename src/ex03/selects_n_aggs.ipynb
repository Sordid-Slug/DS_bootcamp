```python
import pandas as pd
```

## Read JSON

```python
df = pd.read_json('../ex02/auto.json')
df.set_index('CarNumber', inplace=True)
```

## Make selects

```python
df[df['Fines'] > 2100].head()
```

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)].head()
```

```python
df[df['Model'].isin(['Focus', 'Corolla'])].head()
```

```python
car_numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(car_numbers)].head()
```

## Make aggregations with make and model

```python
df.groupby(by='Make').agg({'Fines' : ['median']})
```

```python
df.groupby(by=['Make', 'Model']).agg({'Fines' : ['count']})
```

```python
df.groupby(by=['Make', 'Model']).agg({'Fines' : ['min', 'max']})
```

```python
df.groupby(by=['Make', 'Model']).agg({'Fines' : ['std']})
```

## Make aggregations with the car number

```python
top_fines = df.groupby('CarNumber').agg({'Fines' : ['count']}).sort_values(by=('Fines', 'count'), ascending=False)
top_fines
```

```python
top_car_number = top_fines.index[0]
print(df[df.index == top_car_number])
```

```python
top_paid = df.groupby('CarNumber').agg({'Fines' : ['sum']}).sort_values(by=('Fines', 'sum'), ascending=False)
top_paid
```

```python
car_number_top_paid = top_paid.index[0]
df[df.index == car_number_top_paid]
```

```python
unique_model = df.groupby('CarNumber')['Model'].nunique()
unique_model[unique_model > 1]
```
