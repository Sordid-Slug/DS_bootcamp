```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('../data/checking-logs.sqlite')
```

```python
ab_test = pd.read_csv('../data/ab-test.csv')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8))

ab_test.query("time == 'after'").boxplot(
    column='diff',
    by='group',
    ax=ax[0],
    color=dict(whiskers='red', medians='red', caps='red'),
    boxprops=dict(color='green', linewidth=3),
    medianprops=dict(linewidth=2),
    label='Before'
)
ax[0].set_title('After')

ab_test.query("time == 'before'").boxplot(
    column='diff',
    by='group',
    ax=ax[1],
    color=dict(whiskers='red', medians='red', caps='red'),
    boxprops=dict(color='green', linewidth=3),
    medianprops=dict(linewidth=2),
    label='Before'
)
ax[1].set_title('Before')
```
